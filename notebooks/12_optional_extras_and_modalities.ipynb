# ModSSC | Optional extras and modalities

Inspect the optional dependency profile and map dataset IDs to their required extras.

## Objective
- Read the extras declared in `pyproject.toml`.
- Group curated dataset IDs by `required_extra`.
- Build a small local recommendation map for common workflows.


```python
import tomllib
from pathlib import Path

pyproject = tomllib.loads(Path("pyproject.toml").read_text(encoding="utf-8"))
extras = pyproject["project"]["optional-dependencies"]
sorted(extras)
```

```python
from collections import defaultdict

from modssc.data_loader import available_datasets, dataset_info

by_extra = defaultdict(list)
for key in available_datasets():
    info = dataset_info(key)
    by_extra[info.required_extra or "core"].append(key)

{extra: sorted(keys) for extra, keys in sorted(by_extra.items())}
```

```python
recommendations = {
    "core_cli_api": ["modssc"],
    "most_datasets": ["modssc[datasets]"],
    "preprocess_heavy": ["modssc[preprocess]"],
    "text_inductive": ["modssc[hf,preprocess-text,inductive-torch]"],
    "graph_transductive": ["modssc[graph,transductive-torch]"],
    "full_local_env": ["modssc[full]"],
}
recommendations
```
