# ModSSC | Bench run anatomy

Inspect a completed benchmark run from disk and understand what ModSSC writes.

## Objective
- Locate a recent run directory under `runs/`.
- Read the `run.json` payload.
- Inspect the copied config and the key result blocks.


```python
from pathlib import Path

runs_dir = Path("runs")
run_dirs = sorted([p for p in runs_dir.glob("*") if p.is_dir()])
if not run_dirs:
    raise FileNotFoundError("No run directories found under runs/")
run_dirs[-5:]
```

```python
import json

run_dir = run_dirs[-1]
run = json.loads((run_dir / "run.json").read_text(encoding="utf-8"))
{
    "run_dir": str(run_dir),
    "top_level_keys": sorted(run.keys()),
    "status": run.get("status"),
}
```

```python
copied_config = (run_dir / "config.yaml").read_text(encoding="utf-8")
print(copied_config[:1200])

# Depending on the run type, interesting blocks typically include metrics,
# artifacts, evaluation summaries, and HPO metadata.
{key: run.get(key) for key in ("metrics", "artifacts", "evaluation", "hpo") if key in run}
```
